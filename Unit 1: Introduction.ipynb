# 1’ĖÅŌāŻ Spring Framework

## ­¤ö╣ What is Spring?

Spring is an **open-source Java framework** used to build **enterprise-level applications**.

It helps developers create:

- Loosely coupled applications  
- Easily maintainable systems  
- Scalable enterprise solutions  

Spring manages **object creation and dependency injection automatically**.

---

## ­¤ö╣ Core Concepts of Spring

### Ō£ģ 1. IoC (Inversion of Control)

- Control of object creation is given to the Spring container.
- Instead of creating objects manually using `new`, Spring creates them.

### Ō£ģ 2. DI (Dependency Injection)

- Dependencies are injected automatically by Spring.
- Promotes loose coupling.

Example:
```java
@Autowired
private StudentService service;
```

---

### Ō£ģ 3. AOP (Aspect-Oriented Programming)

- Used to handle cross-cutting concerns.
- Example:
  - Logging
  - Security
  - Transactions

---

### Ō£ģ 4. Spring MVC

- Used to build web applications.
- Follows Model-View-Controller pattern.
- Used for REST APIs and web apps.

# 2’ĖÅŌāŻ Problems with Traditional Spring

Traditional Spring Framework had several issues:

- Manual Bean Configuration
- Large XML Configuration files
- Time-consuming project setup
- Need to install external server (Tomcat)
- Complex dependency management

### ŌØŚ Result:
Development was slow and complex.

---

# 3’ĖÅŌāŻ What is Spring Boot?

Spring Boot is built on top of Spring Framework.

It simplifies Spring development.

### ­¤ö╣ Features of Spring Boot:

- Removes configuration headache (Auto Configuration)
- Runs using embedded Tomcat server
- Automatically connects to database (if configured)
- Used to build REST APIs
- Best for Microservices development

---

# 4’ĖÅŌāŻ Spring vs Spring Boot

| XML-based setup | Minimal configuration |
| External server required | Embedded server |
| Complex dependency management | Starter dependencies |
| More setup time | Fast development |

# 6’ĖÅŌāŻ Auto Configuration

Spring Boot automatically configures the application based on:

- Dependencies present in classpath
- Application properties

### Important Annotations:

```java
@EnableAutoConfiguration
```

```java
@Configuration
```

```java
@SpringBootApplication
```

### What does `@SpringBootApplication` include?

It combines:

- `@Configuration`
- `@EnableAutoConfiguration`
- `@ComponentScan`

---

### ­¤ö╣ Example:

If you add dependency:

```
spring-boot-starter-web
```

Spring Boot automatically configures:

- DispatcherServlet
- Embedded Tomcat
- Jackson (JSON converter)
- Spring MVC

No manual configuration needed.

---

# 7’ĖÅŌāŻ Starter Dependencies

Starter Dependencies are predefined dependency packages provided by Spring Boot.

Instead of adding multiple dependencies, we add one starter.

### Common Starters:

- `spring-boot-starter-web`
- `spring-boot-starter-data-jpa`
- `spring-boot-starter-security`
- `mysql-connector-j`
- Lombok

### Advantage:

Instead of adding 10 dependencies, add 1 starter.

Example:

```xml
<dependency>
    <groupId>org.springframework.boot</groupId>
    <artifactId>spring-boot-starter-web</artifactId>
</dependency>
```

---

# 8’ĖÅŌāŻ Embedded Servers

Spring Boot provides embedded servers:

- Tomcat (Default)
- Jetty
- Undertow

### Advantages:

- No need to install Tomcat separately
- Just run the `main()` method
- Application starts with embedded server

Example:

```java
@SpringBootApplication
public class Application {
    public static void main(String[] args) {
        SpringApplication.run(Application.class, args);
    }
}
```

Application runs on:

```
http://localhost:8080
```
