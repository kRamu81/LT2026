# 1 .What is a Build Tool?

A **Build Tool** is a software tool that automates the process of:

- Compiling source code
- Managing dependencies (external libraries)
- Running tests
- Packaging the application (JAR/WAR)
- Deploying the application

Instead of manually compiling and managing libraries, a build tool handles everything automatically.

###  Why We Need Build Tools?

Without a build tool:
- You must manually download JAR files.
- You must set classpaths manually.
- You must compile files manually.
- Dependency conflicts become difficult to manage.

With a build tool:
- Dependencies are downloaded automatically.
- Project structure is standardized.
- Builds are repeatable and consistent.

# 2 Maven Overview

**Maven** is a popular build automation tool used mainly for Java projects.

- Developed by Apache
- Uses XML configuration file (`pom.xml`)
- Follows Convention over Configuration
- Automatically manages project dependencies

### 🔹 Key Features of Maven

- Dependency Management
- Standard Project Structure
- Lifecycle Management
- Plugin System
- Easy Integration with IDEs

---

# 3 Maven Project Structure

Maven follows a standard directory structure:

```
project-name
│
├── src
│   ├── main
│   │   ├── java
│   │   ├── resources
│   │
│   ├── test
│       ├── java
│       ├── resources
│
├── pom.xml
└── target
```

### 📂 Explanation

- `src/main/java` → Application source code  
- `src/main/resources` → Configuration files (application.properties)  
- `src/test/java` → Test cases  
- `pom.xml` → Project configuration file  
- `target` → Compiled output (generated automatically)

---

# 4 pom.xml Explained

`pom.xml` stands for:

> Project Object Model

It contains project configuration and dependency details.

### Example pom.xml

```xml
<project xmlns="http://maven.apache.org/POM/4.0.0">
    
    <modelVersion>4.0.0</modelVersion>

    <groupId>com.ramu</groupId>
    <artifactId>student-project</artifactId>
    <version>0.0.1-SNAPSHOT</version>
    <packaging>jar</packaging>

    <name>Student Project</name>
    <description>Spring Boot Layered Architecture</description>

    <dependencies>
        <!-- Spring Boot Web Dependency -->
        <dependency>
            <groupId>org.springframework.boot</groupId>
            <artifactId>spring-boot-starter-web</artifactId>
        </dependency>

        <!-- Spring Data JPA -->
        <dependency>
            <groupId>org.springframework.boot</groupId>
            <artifactId>spring-boot-starter-data-jpa</artifactId>
        </dependency>

        <!-- MySQL Driver -->
        <dependency>
            <groupId>com.mysql</groupId>
            <artifactId>mysql-connector-j</artifactId>
        </dependency>
    </dependencies>

</project>
```

### 🔹 Important Tags

- `<groupId>` → Organization name
- `<artifactId>` → Project name
- `<version>` → Project version
- `<dependencies>` → External libraries
- `<packaging>` → jar / war

# 5 Dependencies and Plugins

## 🔹 Dependencies

Dependencies are external libraries required for the project.

Example:
- Spring Boot
- MySQL Driver
- Lombok
- JUnit

Maven automatically:
- Downloads from Maven Central Repository
- Stores in local repository (`.m2` folder)

---

## 🔹 Plugins

Plugins perform specific tasks during the build process.

Example:

```xml
<build>
    <plugins>
        <plugin>
            <groupId>org.springframework.boot</groupId>
            <artifactId>spring-boot-maven-plugin</artifactId>
        </plugin>
    </plugins>
</build>
```

### Common Plugins

- Compiler Plugin
- Surefire Plugin (Testing)
- Spring Boot Maven Plugin
- WAR Plugin

# 6️⃣ Maven Lifecycle

Maven has 3 main lifecycles:

### 1️⃣ Clean Lifecycle
- `mvn clean`
- Deletes target folder

### 2️⃣ Default Lifecycle
Main build lifecycle:

- `validate`
- `compile`
- `test`
- `package`
- `verify`
- `install`
- `deploy`

### Example Commands:

```
mvn compile
mvn test
mvn package
mvn install
```

### 3️⃣ Site Lifecycle
- Generates project documentation

# 7️⃣ What is Gradle?

**Gradle** is another modern build automation tool.

- Uses Groovy or Kotlin DSL
- Faster than Maven
- More flexible
- Less verbose configuration

Example Gradle file:

```groovy
plugins {
    id 'java'
    id 'org.springframework.boot' version '3.2.0'
}

dependencies {
    implementation 'org.springframework.boot:spring-boot-starter-web'
    implementation 'org.springframework.boot:spring-boot-starter-data-jpa'
    runtimeOnly 'com.mysql:mysql-connector-j'
}
```

---

# 8️⃣ Difference Between Maven and Gradle

| Feature | Maven | Gradle |
|----------|--------|---------|
| Configuration | XML | Groovy/Kotlin DSL |
| Performance | Slower | Faster |
| Flexibility | Less flexible | Highly flexible |
| Learning Curve | Easier | Moderate |
| Build Scripts | pom.xml | build.gradle |
| Convention | Strong conventions | More customizable |

- Build tools automate compilation and dependency management.

- Gradle is faster and more flexible.
